```python
# Import Dependencies 
import pandas as pd
from tabula import read_pdf
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read In CSV
df = pd.read_csv('gun-violence-data_01-2013_03-2018.csv')
```

```python
# Create Specific Gun Violence Data Frame
df['year'] = pd.DatetimeIndex(df['date']).year
df2 = df.loc[:,['year', 'date', 'state', 'n_killed', 'n_injured']]
```

```python
# Create 2014 Data
df2014 = df2.loc[df2.year==2014]
number_killed = df2014['n_killed'].sum()                  

# Create Number Killed 2014 Data Frame
killed2014 = pd.DataFrame([{"Killed 2014": number_killed}]) 

# Create Total Gun Violence Cases 2014 Data Frame
total_v_2014 = df2014["year"].count()
violence2014 = pd.DataFrame([{"Total": total_v_2014}]) 
Year = ['2014']
violence2014['Year'] = Year
violence2014
```

```python
# Create 2015 Data
df2015 = df2.loc[df2.year==2015]

# Create Number Killed 2015 Data Frame
number_killed = df2015['n_killed'].sum()                  
killed2015 = pd.DataFrame([{"Killed 2015": number_killed}])      

# Create Total Gun Violence Cases 2015 Data Frame
total_v_2015 = df2015["year"].count()
violence2015 = pd.DataFrame([{"Total": total_v_2015}]) 
Year = ['2015']
violence2015['Year'] = Year
v2014_2015 = violence2014.merge(violence2015, how='outer')
v2014_2015
```

```python
# Create 2016 Data
df2016 = df2.loc[df2.year==2016]

# Create Number Killed 2016 Data Frame
number_killed = df2016['n_killed'].sum()                  
killed2016 = pd.DataFrame([{"Killed 2016": number_killed}])      # Data Frame Number Killed in 2016

# Create Total Gun Violence Cases 2016 Data Frame
total_v_2016 = df2016["year"].count()
violence2016 = pd.DataFrame([{"Total": total_v_2016}]) 
Year = ['2016']
violence2016['Year'] = Year
violence2016
```

```python
# Create 2017 Data
df2017 = df2.loc[df2.year==2017]

# Create Number Killed 2017 Data Frame
number_killed = df2017['n_killed'].sum()                  
killed2017 = pd.DataFrame([{"Killed 2017": number_killed}])      # Data Frame Number Killed in 2017

# Create Total Gun Violence Cases 2017 Data Frame
total_v_2017 = df2017["year"].count()
violence2017 = pd.DataFrame([{"Total": total_v_2017}]) 

# Create Data Frame For 2017 And Data Frame For Bar Graph
Year = ['2017']
violence2017['Year'] = Year
v2016_2017 = violence2016.merge(violence2017, how='outer')
Total_df = v2014_2015.merge(v2016_2017, how='outer')
Total_df
```

```python
# Create Data Frame For Bar Graph for Gun Killings Year to Year
combined1 = killed2014.join(killed2015)
combined2 = combined1.join(killed2016)
combined3 = combined2.join(killed2017)
combined3.plot(kind='barh', yticks=combined3.columns)
# plt.xticks(np.arange(len(lst)), lst)
plt.xlabel("People Killed", fontsize=13)
plt.ylabel("Year", fontsize=13)
plt.title("American Gun Violance Killings Per Year", fontsize=15)
plt.grid()
```

```python
# Create Bar Graph For Total Gun Violance Per Year
x_axis = np.arange(len(Total_df))
tick_locations = [value+0.4 for value in x_axis]
plt.bar(x_axis, Total_df["Total"], alpha=0.5, align="edge", color="grbc")
plt.xticks(tick_locations, Total_df["Year"], rotation="vertical")
plt.grid()
plt.rc('axes', axisbelow=False)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Gun Violance Incidents', fontsize=15)
plt.title("Gun Violance Incidents(2014-2017)", fontsize=18)
plt.show()
```
