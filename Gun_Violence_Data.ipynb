# __Gun Violence__

```python
# Import Dependencies 
import pandas as pd
from tabula import read_pdf
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read In CSV
df = pd.read_csv('gun-violence-data_01-2013_03-2018.csv')
population = pd.read_csv('population.csv')
manufactured = pd.read_csv('Firearms Manufactured (2014-2017).csv')
population_state = pd.read_csv('population-2014-2017.csv')
```

# Create Gun Violence Data Frame 

```python
# Create Specific Gun Violence Data Frame
df['year'] = pd.DatetimeIndex(df['date']).year
df2 = df.loc[:,['year', 'date', 'state', 'n_killed', 'n_injured']]
df3 = df2.loc[:,['year', 'state']]
```

# 2014 Data

```python
# Create 2014 Data
df2014 = df2.loc[df2.year==2014]                  

# Create Number Killed 2014 Data Frame
Year = ['2014']
number_killed = df2014['n_killed'].sum()
killed2014 = pd.DataFrame([{"Killed": number_killed}])
killed2014['Year'] = Year

# Create Total Gun Violence Cases 2014 Data Frame
total_v_2014 = df2014["year"].count()
violence2014 = pd.DataFrame([{"Total": total_v_2014}]) 
violence2014['Year'] = Year
```

# 2015 Data

```python
# Create 2015 Data
df2015 = df2.loc[df2.year==2015]

# Create Number Killed 2015 Data Frame
Year = ['2015']
number_killed = df2015['n_killed'].sum()                  
killed2015 = pd.DataFrame([{"Killed": number_killed}])
killed2015['Year'] = Year
ktotal = killed2014.merge(killed2015, how='outer')

# Create Total Gun Violence Cases 2015 Data Frame
total_v_2015 = df2015["year"].count()
violence2015 = pd.DataFrame([{"Total": total_v_2015}]) 
violence2015['Year'] = Year
v2014_2015 = violence2014.merge(violence2015, how='outer')
v2014_2015
```

# 2016 Data

```python
# Create 2016 Data
df2016 = df2.loc[df2.year==2016]

# Create Number Killed 2016 Data Frame
Year = ['2016']
number_killed = df2016['n_killed'].sum()                  
killed2016 = pd.DataFrame([{"Killed": number_killed}])      
killed2016['Year'] = Year

# Create Total Gun Violence Cases 2016 Data Frame
total_v_2016 = df2016["year"].count()
violence2016 = pd.DataFrame([{"Total": total_v_2016}]) 
violence2016['Year'] = Year
```

# 2017 Data

```python
# Create 2017 Data
df2017 = df2.loc[df2.year==2017]

# Create Number Killed 2017 Data Frame
Year = ['2017']
number_killed = df2017['n_killed'].sum()                  
killed2017 = pd.DataFrame([{"Killed": number_killed}])
killed2017['Year'] = Year
k2 = killed2016.merge(killed2017, how='outer')
k3 = ktotal.merge(k2, how='outer')

# Create Total Gun Violence Cases 2017 Data Frame
total_v_2017 = df2017["year"].count()
violence2017 = pd.DataFrame([{"Total": total_v_2017}])

# Create Data Frame For 2017 And Data Frame For Bar Graph
violence2017['Year'] = Year
v2016_2017 = violence2016.merge(violence2017, how='outer')
Total_df = v2014_2015.merge(v2016_2017, how='outer')
Total_df
```

# Gun Killings Data

```python
# Create Data Frame For Bar Graph for Gun Killings Year to Year
x_axis = np.arange(len(k3))
tick_locations = [value+0.4 for value in x_axis]
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plt.bar(x_axis, k3["Killed"], alpha=0.8, align="edge", color=colors)
plt.xticks(tick_locations, k3["Year"], rotation="vertical")
plt.xlabel("People Killed", fontsize=13)
plt.ylabel("Year", fontsize=13)
plt.title("American Gun Killings Per Year", fontsize=15)
plt.grid()

# Save the Figure
plt.savefig('Killings_per_year.png')

# Display Figure
plt.show()
```

# Create Total Gun Violence Incidents (2014-2017)

```python
# Create Bar Graph For Total Gun Violance Per Year
x_axis = np.arange(len(Total_df))
tick_locations = [value+0.4 for value in x_axis]
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plt.bar(x_axis, Total_df["Total"], alpha=0.8, align="edge", color=colors)
plt.xticks(tick_locations, Total_df["Year"], rotation="vertical")
plt.grid()
plt.rc('axes', axisbelow=False)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Gun Violence Incidents', fontsize=15)
plt.title("Gun Violence Incidents(2014-2017)", fontsize=18)

# Save the Figure
plt.savefig('Gun_Violence_Incidents.png')

# Display Figure
plt.show()
```

# Create Total Population Data (2014-2017)

```python
# Create Data Frame Population 2014-2017
population1 = population.dropna(how='all')
pop2 = population1.rename(columns={'Unnamed: 1':'Population'})
pop3 = pop2.rename(columns={'United States resident population 2014-2017':'Year'})
pop3['Population'] = pd.to_numeric(pop3["Population"].str.replace(',',''))

# Create Bar Graph For Total Population Per Year
x_axis = np.arange(len(pop3))
tick_locations = [value+0.4 for value in x_axis]
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plt.bar(x_axis, pop3["Population"], alpha=0.8, align="edge", color=colors)
plt.xticks(tick_locations, pop3["Year"], rotation="vertical")
plt.ylim([min(pop3['Population']) - 1000, max(pop3['Population']) + 1000])
plt.rc('axes', axisbelow=False)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Population', fontsize=15)
plt.title("Total Population(2014-2017)", fontsize=18)

# Save the Figure
plt.savefig('Total_Population.png')

# Display Figure
plt.show()
```

# Create Total Guns Manufactured Data (2014-2017)

```python
# Create Specific Data Frame for Manufactured Data
manufactured1 = manufactured.loc[:,['Year', 'Total_Firearms']]
manufactured2 = manufactured1.dropna(how='all')
manufactured2['Total_Firearms'] = pd.to_numeric(manufactured2['Total_Firearms'].str.replace(',',''))

# Create Bar Graph For Manufactured Data Per Year
x_axis = np.arange(len(manufactured2))
tick_locations = [value+0.4 for value in x_axis]
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plt.bar(x_axis, manufactured2["Total_Firearms"], alpha=0.8, align="edge", color=colors)
plt.xticks(tick_locations, manufactured2["Year"], rotation="vertical")
plt.grid()
plt.rc('axes', axisbelow=False)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Total Firearms', fontsize=15)
plt.title("Toatl Firearms Manufactured(2014-2017)", fontsize=14)

# Save the Figure
plt.savefig('Total_Firearms.png')

# Display Figrue
plt.show()
```

# Create Total Gun Violence Cases Data

```python
# Create Total Gun Violence Cases Data Frame
Countdf = df2[(df2.year >2013)]
countdf2 = Countdf.groupby(['state'])['year'].count()/3
countdf3 = pd.DataFrame(countdf2)
countdf4 = countdf3.reset_index()

# Create Total Gun Violence Cases Bar Graph
x_axis = np.arange(len(countdf4))
tick_locations = [value+0.4 for value in x_axis]
plt.figure(figsize=(20,10))
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plt.bar(x_axis, countdf4['year'], align="edge", color=colors)
plt.xticks(tick_locations, countdf4["state"], rotation="vertical", fontsize=(20))
plt.yticks(fontsize=(20))
plt.grid()
plt.title('Average Total Gun Violence Cases Per State(2014-2017)',fontsize=(39))
plt.ylabel('Total Gun Violence Cases',fontsize=(32))

# Save the Figure
plt.savefig('Average_Total_GunV_per_state.png')

# Display Figrue
plt.show()
```

```python
# Create Data Frame For Pie Chart
countdf5 = countdf4.sort_values(by='year',ascending=False)
countdf6 = countdf5.head(7)

# Create Top 7 Total Gun Violence Per State Pie Chart
fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111)
plt.rcParams['font.size'] = 27.0
cmap = plt.cm.prism
labels = countdf6["state"]
explode = (.2,0,0,0,0,0,0)
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
ax.pie(countdf6['year'], labels=labels, shadow=True, colors=colors, 
       explode=explode, labeldistance=1.05,autopct='%1.1f%%')
ax.set_title("Top 7 Total Gun Violence Cases Per State");

# Save the Figure
plt.savefig('Total_GunV_Top7_Pie.png')

# Display Figrue
plt.show()
```

# Create Population By State Information

```python
# Create Population By State Date Frame
popstate1 = population_state.dropna(how='any')
popstate2 = popstate1.rename(columns={'table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)': 'State',
                                      'Unnamed: 3': '2010', 'Unnamed: 4': '2011', 'Unnamed: 5': '2012',
                                      'Unnamed: 6': '2013', 'Unnamed: 7': '2014', "Unnamed: 8": '2015',
                                      'Unnamed: 9': '2016', 'Unnamed: 10': '2017'})
popstate3 = popstate2.loc[:,['State','2017']]
popstate4 = popstate3[5:]
popstate4['2017'] = pd.to_numeric(popstate4['2017'].str.replace(',',''))

# Create Population By State Bar Graph
x_axis = np.arange(len(popstate4))
tick_locations = [value+0.4 for value in x_axis]
plt.figure(figsize=(40,20))
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plt.bar(x_axis, popstate4['2017'], align="edge", color=colors)
plt.xticks(tick_locations, popstate4["State"], rotation="vertical", fontsize=(38))
plt.grid()
plt.title('Total Population Per State(2017)',fontsize=(70))
plt.ylabel('Number of People By Million',fontsize=(50))
plt.yticks(fontsize=(45))

# Save the Figure
plt.savefig('Total_Population_Per_State.png')

# Display Figrue
plt.show()
```

```python
# Create Data Frame For Pie Chart
popstate5 = popstate4.sort_values(by='2017',ascending=False)
popstate6 = popstate5.head(7)

# Create Top 7 Total Gun Violence Per State Pie Chart
fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111)
plt.rcParams['font.size'] = 27.0
cmap = plt.cm.prism
labels = popstate6["State"]
explode = (.2,0,0,0,0,0,0)
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
ax.pie(popstate6['2017'], explode=explode, colors=colors, 
       shadow=True, labels=labels, labeldistance=1.05,autopct='%1.1f%%')
ax.set_title("Top 7 Total Population By State (2017)");

# Save the Figure
plt.savefig('Total_Pop_Top7_Pie.png')

# Display Figrue
plt.show()
```
